### HTML

###### Websites are written in HTML, so when you pull information directly from a site, you will get all the code back along with the text

```python
import urllib.request
```

```python
url = "https://en.wikipedia.org/wiki/Python_(programming_language)"
```

```python
response = urllib.request.urlopen(url)
```

```python
html = response.read()
```

```python
html[:1000]
```

We will use a Python library called BeautifulSoup to make working with the HTML easier.

```python
from bs4 import BeautifulSoup
```

```python
bs = BeautifulSoup(html,"html5lib")
bs
```

Wikipedia places most readable text within paragraph tags or "p" tags.

```python
web_para = [p_tag.text for p_tag in bs.find_all('p')]
web_para
```

```python
web_tokens = [nltk.word_tokenize(paragraph) for paragraph in web_para]
```

Now that we have each paragraph tokenized, we can find the first one on the page.

```python
print(web_tokens[2])
```

### CSV

Importing text stored in cells on a spreadsheet or in a csv file (comma separated values)

```python
import csv
```

```python
comments = []
with open("reviews.csv", 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        comments.append(row)
```

```python
comments
```

```python
len(comments)
```

This command will give us a list, where each element is another list of all the tokens for each comment.

```python
tokens = [nltk.word_tokenize(str(entry)) for entry in comments]
```

```python
tokens
```

```python
tokens[0]
```

```python
len(tokens)
```

### Remove Stopwords

```python
alice = nltk.corpus.gutenberg.words("carroll-alice.txt")
```

```python
alice_fd = nltk.FreqDist(alice)
```

```python
alice_100 = alice_fd.most_common(100)
```

```python
alice_100
```

```python
alice_common = [word[0] for word in alice_100]
```

```python
alice_common
```

```python
common = set(word.lower() for word in alice_common if word.isalpha())
```

```python
non_stopwrds = list(set(common) - set(nltk.corpus.stopwords.words('english')))
```
